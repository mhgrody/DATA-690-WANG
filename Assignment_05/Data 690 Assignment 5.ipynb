```python
#Create smaller dataframe with only selected rows in keepList

keepList = ['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']

dropList = [x for x in df.columns if x not in keepList]

dfSmall = df.drop(dropList, axis=1)
```

# Variable Definitions:

### UNITID: Unit ID for institution
### INSTNM: Insitution Name
### CITY: City
### STABBR: State postcode
### ZIP: Zip code
### ADM_RATE: Admission rate
### UGDS: Enrollment of undergraduate certificate/degree-seeking students
### TUITIONFEE_IN: In-state tution and fees
### MN_EARN_WNE_P10: Mean earnings of students working and not enrolled 10 years after entry
