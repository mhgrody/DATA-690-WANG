```python
#read in text file line by line
f = open("census_cost.txt", "r")
line_list = [x for x in f]
```

```python
print(line_list)
```

```python
for line in line_list:
    print(line)
```

```python
#top 2 lines
top2_list = line_list[:2]
top2_list
```

```python
#rest of the data
data_list = line_list[2:]
print(data_list)
```

```python
#extract 'Census Year' column
split_data = [x.split('\t') for x in data_list]
split_data
```

```python
#list of years, removing the * from last value
year_list = [x[0][:4] for x in split_data]
print(year_list)
```

```python
#list of years, removing the ','
pop_list = [x[1].replace(',', '') for x in split_data]
print(pop_list)
```

```python
#list of cost, cleaning up the values
cost_list = [x[2].replace(',', '').replace('$', '').replace(' Billion', '000000000').replace('.', '')
             for x in split_data]
print(cost_list)
```

```python
#avg list and cleaning
avg_list = []

for line in split_data:
    if line[-1][-2] == 's':
        avg_list.append(str(float(line[-1].replace(' cents\n', ''))/100))
    else:
        avg_list.append(line[-1].replace('$', '').replace('\n', ''))
        
print(avg_list)
```

```python
#create list of each row of cleaned data
cleaned_lines = []
for i in range(len(avg_list)):
    s1 = year_list[i]+ ',' + pop_list[i] + ',' + cost_list[i] + ',' + avg_list[i] 

    cleaned_lines.append(s1)
cleaned_lines
```

```python
for i in range(len(top2_list)):
    top2_list[i] = top2_list[i].replace('\n', '').replace('\t', ',')

end_file = top2_list + cleaned_lines
end_file
```

```python
outF = open("census_cost.csv", "w")
for line in end_file:
  # write line to output file
    outF.write(line)
    outF.write("\n")
outF.close()
```
